```python
dfs["Nutriscore"] = df_clean["Nutriscore"].str.extract(r"([A-E])")
```

```python
dfs.head()
```

```python
import re
import numpy as np
def clean_value_kj(x):
    # Si NaN → retourner NaN
    if pd.isna(x):
        return np.nan

    # Si pas une chaîne → garder tel quel
    if not isinstance(x, str):
        return x

    # Nettoyage de base
    x = x.replace(",", ".").strip()

    # Gestion du "?"
    if x == "?" or x == "":
        return np.nan

    # Extraire tous les nombres présents
    numbers = re.findall(r"[-+]?\d*\.?\d+", x)

    # Si aucun nombre → NaN
    if not numbers:
        return np.nan

    # Retourner le PREMIER nombre = celui en kJ
    return float(numbers[0])
```

```python
nutrition_cols = ["énergie", "acides_gras_saturés", "sucres", "protéines", "sel","fibres_alimentaires"]  # adapte selon ton dataset

for col in nutrition_cols:
    dfs[col] = dfs[col].apply(clean_value_kj)
```
